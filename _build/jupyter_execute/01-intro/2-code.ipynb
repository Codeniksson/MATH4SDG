Lets from 

$$
 y = \sin (x), \quad x \in [0, 10 \pi]
$$



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10*np.pi, 500)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave Example')
plt.show()
```
